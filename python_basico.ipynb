## 2. Listas y Tuplas

### Listas
- Equivalente a arrays/celdas en MATLAB
- Se definen con corchetes `[]`
- Son **mutables** (se pueden modificar)

````python
# Crear listas
ciudades = ['NYC', 'LA', 'Chicago', 'Houston']
numeros = [1, 2, 3, 4, 5]
mixta = [1, 'dos', 3.0, True]  # Python permite tipos mixtos

print("Primera ciudad:", ciudades[0])      # ndice 0 = primer elemento
print("ltima ciudad:", ciudades[-1])      # ndice -1 = 煤ltimo elemento
print("Segunda y tercera:", ciudades[1:3]) # Slicing [inicio:fin] (fin no incluido)
````

````python
# Operaciones con listas
almacenes = ['Harlingen', 'Memphis', 'Ashland']

# A帽adir elementos
almacenes.append('Dallas')      # A帽ade al final
print("Despu茅s de append:", almacenes)

# Longitud de la lista
n = len(almacenes)              # MATLAB: length()
print(f"N煤mero de almacenes: {n}")

# Verificar si un elemento existe
if 'Memphis' in almacenes:
    print("Memphis est谩 en la lista")
````

## 3. Diccionarios

**隆MUY IMPORTANTE PARA PYOMO!**

Los diccionarios almacenan pares **key-value**. Son equivalentes a `struct` en MATLAB.

- Se definen con llaves `{}`
- Formato: `{key1: value1, key2: value2}`
- Las claves pueden ser strings, n煤meros o tuplas

````python
# Diccionario simple: componente -> coste
costes = {
    'A': 0.32,
    'B': 0.25,
    'C': 0.05
}

print("Coste de A:", costes['A'])
print("Coste de B:", costes['B'])

# A帽adir nuevo elemento
costes['D'] = 0.40
print("\nDiccionario actualizado:", costes)
````

````python
# Diccionario con tuplas como keys
# Representa costes de transporte entre almac茅n y ciudad
coste_transporte = {
    ('Harlingen', 'NYC'): 1956,
    ('Harlingen', 'LA'): 1606,
    ('Memphis', 'NYC'): 1096,
    ('Memphis', 'LA'): 1792
}

# Acceder a un valor espec铆fico
print("Coste Harlingen -> NYC:", coste_transporte[('Harlingen', 'NYC')])

# Tambi茅n se puede escribir sin el par茅ntesis extra:
print("Coste Memphis -> LA:", coste_transporte['Memphis', 'LA'])
````

````python
# Iterar sobre diccionarios
print("Todas las rutas y sus costes:")
for ruta, coste in coste_transporte.items():
    almacen, ciudad = ruta  # Desempaquetar tupla
    print(f"  {almacen} -> {ciudad}: ${coste}")
````

## 4. Bucles y Comprensiones

### Bucle `for`
En Python, `for` itera directamente sobre elementos (diferente a MATLAB donde iteras sobre 铆ndices)

````python
# Iterar sobre una lista
componentes = ['A', 'B', 'C']

print("Forma Python (iterar sobre elementos):")
for comp in componentes:
    print(f"Componente: {comp}")

print("\nSi necesitas 铆ndices, usa enumerate():")
for i, comp in enumerate(componentes):
    print(f"ndice {i}: {comp}")
````

````python
# range() para generar secuencias de n煤meros
# MATLAB: for i = 1:5

print("range(5) genera: 0, 1, 2, 3, 4")
for i in range(5):
    print(i, end=' ')

print("\n\nrange(1, 6) genera: 1, 2, 3, 4, 5")
for i in range(1, 6):
    print(i, end=' ')
````

### List comprehension

Permiten crear listas de forma concisa. Muy usadas en restricciones y funciones objetivo de Pyomo.

````python
# Forma tradicional
cuadrados = []
for i in range(5):
    cuadrados.append(i**2)
print("Forma tradicional:", cuadrados)

# Comprensi贸n de lista (m谩s compacta)
cuadrados = [i**2 for i in range(5)]
print("Comprensi贸n de lista:", cuadrados)
````

````python
# Ejemplos relevantes para Pyomo
componentes = ['A', 'B', 'C']
costes = {'A': 2, 'B': 2, 'C': 5}

# Crear lista de costes
lista_costes = [costes[c] for c in componentes]
print("Lista de costes:", lista_costes)

# Con filtro (if)
componentes_baratos = [c for c in componentes if costes[c] < 3]
print("Componentes baratos:", componentes_baratos)
````

### Funci贸n sum()

````python
# Sumar elementos de una lista
numeros = [1, 2, 3, 4, 5]
total = sum(numeros)
print(f"Suma de {numeros} = {total}")

# Combinar sum() con comprensi贸n de lista (patr贸n t铆pico en Pyomo)
componentes = ['A', 'B', 'C']
cantidades = {'A': 0.3, 'B': 0.5, 'C': 0.2}
costes = {'A': 2, 'B': 2, 'C': 5}

# Calcular coste total: sum(cantidad[c] * coste[c] for c in componentes)
coste_total = sum(cantidades[c] * costes[c] for c in componentes)
print(f"\nCoste total: {coste_total}")
````

## 5. Funciones

### Funciones regulares

````python
# Definir una funci贸n
def calcular_coste(cantidad, precio):
    """Calcula el coste total dado cantidad y precio."""
    return cantidad * precio

# Usar la funci贸n
coste = calcular_coste(100, 0.25)
print(f"Coste: ${coste}")
````

````python
# Funci贸n con m煤ltiples valores de retorno
def estadisticas(numeros):
    """Retorna suma y promedio de una lista."""
    total = sum(numeros)
    promedio = total / len(numeros)
    return total, promedio  # Retorna tupla

datos = [10, 20, 30, 40]
suma, prom = estadisticas(datos)
print(f"Suma: {suma}, Promedio: {prom}")
````

### Funciones Lambda
Permiten definir restricciones simples de forma compacta

````python
# Funci贸n regular
def cuadrado(x):
    return x**2

# Funci贸n lambda equivalente (an贸nima, de una l铆nea)
cuadrado_lambda = lambda x: x**2

print("Funci贸n regular:", cuadrado(5))
print("Funci贸n lambda:", cuadrado_lambda(5))
````

````python
# Lambdas con m煤ltiples argumentos
suma = lambda a, b: a + b
print("Suma:", suma(10, 20))
````
